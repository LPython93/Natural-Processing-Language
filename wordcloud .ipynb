# Wikipedia - Wordcloud

```python
import wikipedia
from wordcloud import WordCloud
import glymur
import PIL
import os
from os import path
from PIL import Image
import numpy as np
%matplotlib inline
```

```python
# wikipedia
def get_wiki(query):
    
    title = wikipedia.search(query)[0]
    page = wikipedia.page(title)
    return page.content

# wordcloud
wordcloud = WordCloud().generate(get_wiki("cristiano ronaldo"))
import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

wordcloud = WordCloud(max_font_size=40).generate(get_wiki("cristiano ronaldo"))
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
